```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../task1/ml_approach/data/train_basic.csv')
```

```python
df = df.dropna(subset = ['power', 'price', 'depreciation', 'age'])
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
x = df['power']
y = df['depreciation']
ax1.set_title('Power vs dep')
ax1.scatter(x, y)
x = df['price']
y = df['depreciation']
ax2.set_title('Price vs depn')
ax2.scatter(x, y)
x = df['age']
y = df['depreciation']
ax3.set_title('Age vs dep')
ax3.scatter(x, y)
```

```python
sports = df.loc[df['sports car'] == 1, 'depreciation'].dropna()
truck = df.loc[df['truck'] == 1, 'depreciation'].dropna()
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.set_title('Sports car depreciation')
ax1.boxplot(sports)
ax2.set_title('Truck depreciation')
ax2.boxplot(truck)
```

```python
normal = df.loc[df['electric'] == 0, 'depreciation'].dropna()
electric = df.loc[df['electric'] == 1, 'depreciation'].dropna()
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.set_title('Normal fuel depreciation')
ax1.boxplot(sports)
ax2.set_title('Electric depreciation')
ax2.boxplot(truck)
```
