```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
train = pd.read_csv('calculated_dereg.csv')
test = pd.read_csv('calculated_dereg_test.csv')
df = pd.concat([train, test])

# #
# useful = ['model', 'age', 'mileage']
# df = df.dropna(subset = useful)
# df.info()
```

```python
# get the missing depreciations
dfv = df[~df.depreciation.isnull()]
dfm = df[df.depreciation.isnull()]
search = []
for idx, row in dfm.iterrows():
    model = row['model']
    age = row['age'] // 12 * 12
    mam = (model, age)
    search.append(mam)

from collections import defaultdict
mapper = defaultdict(list)
for idx, row in dfv.iterrows():
    model = row['model']
    age = row['age'] // 12 * 12
    mam = (model, age)
    if mam in search:
        mapper[mam].append(row['depreciation'])

for idx, row in dfm.iterrows():
    model = row['model']
    age = row['age'] // 12 * 12
    mam = (model, age)
    avg = np.mean(mapper[mam])
    df.loc[idx, 'depreciation'] = avg

print(df['depreciation'].isnull().sum())
```
